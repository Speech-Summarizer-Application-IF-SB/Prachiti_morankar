```python
import soundfile as sf
import noisereduce as nr

def denoise_audio(input_file, output_file="cleaned.wav"):
    # Read audio file
    data, rate = sf.read(input_file)

    # Take first 0.5 seconds as noise sample
    noise_clip = data[:int(0.5 * rate)]

    # Reduce noise
    cleaned = nr.reduce_noise(y=data, sr=rate, y_noise=noise_clip)

    # Save cleaned audio
    sf.write(output_file, cleaned, rate)
    print(f"ðŸŽ§ Cleaned audio saved as {output_file}")
    return output_file

out = denoise_audio(file_name, "denoised.wav")
```
